```python
# renaming columns and saving to df_clean

df_clean.rename(columns=dict, inplace=True)
```

```python
# confirm that new column names are changed

display(df_clean)
```

```python
# adding hyphen to separate year and month

df_clean['Year_Month'] = df_clean['Year_Month'].str.replace(r'(\w{4})',r'\1-',regex=True).str.strip('\-')
```

```python
df_clean.head()
```

```python
# separating month from Year_Month values and creating 'Month' column

df_clean['Month'] = pd.DatetimeIndex(df_clean['Year_Month']).month
df_clean.head()
```

```python
# separating year from Year_Month values and creating 'Year' column

df_clean['Year'] = pd.DatetimeIndex(df_clean['Year_Month']).year
df_clean.head()
```

```python
# defining a list of columns

cols = df_clean.columns.tolist()
cols
```

```python
# moving Year and Month column to beginning of dataframe

cols = cols[-2:] + cols[:-2]
cols
```

```python
# swapping position of Month and Year 

cols[0],cols[2] = cols[2],cols[0]
cols
```

```python
# assigning new column positions to data frame and verifying changes

df_clean = df_clean[cols]
df_clean
```

```python
# drop column Year_Month

df_clean = df_clean.drop(columns = ['Year_Month'])
```

```python
# searching for columns with mixed type data. none found.

for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].map(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```
